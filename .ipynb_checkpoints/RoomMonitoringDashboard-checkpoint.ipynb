```python
from jupyter_dash import JupyterDash
from dash import dcc, html
import plotly.express as px
import pandas as pd
```

```python
df = pd.DataFrame({
    "Frutas": ["Maçãs", "Laranjas", "Bananas", "Maçãs", "Laranjas", "Bananas"],
    "Qtd": [4, 1, 2, 2, 4, 5],
    "UF": ["RJ", "RJ", "RJ", "SP", "SP", "SP"]
})

df
```

```python
ss = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
app = JupyterDash(__name__,external_stylesheets=ss)

colors = {'background': '#111111',
         'text': '#7FDBFF'}

fig = px.bar(df, x="Frutas", y="Qtd", color="UF", barmode='group')
fig.update_layout(plot_bgcolor='#111111',
                 paper_bgcolor='#111111',
                 font_color='#7FDBFF')

app.layout = html.Div(style={'backgroundColor': '#111111'}, 
                      children =[html.H1(children='Oi Dash!',
                                        style={'textAlign': 'center',
                                              'color': '#7FDBFF'}),

                                html.Div(children='''
                                Dash para aplicativos web com Python''',
                                        style={'textAlign': 'center',
                                              'color': '#7FDBFF'}),

                                dcc.Graph(
                                    id='example-graph',
                                    figure=fig)                                 
                                ])
```

```python
app.run_server(mode='external')
```

# Back to creating dashboards

```python
df = pd.read_csv('https://gist.githubusercontent.com/chriddyp/c78bf172206ce24f77d6363a2d754b59/raw/c353e8ef842413cae56ae3920b8fd78468aa4cb2/usa-agricultural-exports-2011.csv')
df
```

```python
def generate_table(dataframe, max_rows=10):
  return html.Table([
        html.Thead(
            html.Tr([html.Th(col) for col in dataframe])
        ),
        html.Tbody([
            html.Tr([
                html.Td(dataframe.iloc[i][col]) for col in dataframe.columns
              ]) for i in range(min(len(dataframe),max_rows))            
        ])
  ])
```

```python
app.layout = html.Div(children=[ 
    html.H4(children='Exportações Agrícolas dos EUA (2011)'),
    generate_table(df,20)                            
])


app.run_server(mode='external')
```

# Using Plotly to build graphs

```python
# Scatter Plot

df = pd.read_csv('https://gist.githubusercontent.com/chriddyp/5d1ea79569ed194d432e56108a04d188/raw/a9f9e8076b837d541398e999dcbac2b2826a81f8/gdp-life-exp-2007.csv')

fig = px.scatter(df, x='gdp per capita', y='life expectancy', 
                 size='population',
                color='continent',
                hover_name='country',
                log_x=True,
                size_max=60)

app.layout = html.Div(dcc.Graph(id="T_id", figure=fig))

app.run_server()
```

# Widgets on Plotly graphs

```python
# Dropdown Menu

# Options (arg de dcc.Dropdown) pode ser um dicionário com keys=["label", "value"]
app.layout = html.Div([html.Label("DropdownLabel"),
                      dcc.Dropdown(options=["a", "b", "c"],
                                   value="MLT")],
                     style={"columnCount":1})

app.run_server()
```

```python
# Multiselection Dropdown
# Options (arg de dcc.Dropdown) pode ser um dicionário com keys=["label", "value"]
app.layout = html.Div([html.Label("Dropdown Label"),
                      dcc.Dropdown(options=["a", "b", "c"],
                                   value=["MLT", "SF"],
                                   multi=True)],
                     style={"columnCount":1})

app.run_server()
```

```python
# Radioitens
app.layout = html.Div([html.Label("Radio Items"),
                      dcc.RadioItems(options=["a", "b", "c"],
                                    value="MLT")],
                     style={"columnCount":1})

app.run_server()
```

```python
# Input
app.layout = html.Div([html.Label("Input"),
                      dcc.Input(value="Test", type="text")],
                     style={"columnCount":1})

app.run_server()
```

```python
# Slider
app.layout = html.Div([html.Label("Slider"),
                      dcc.Slider(0, 10, 1, value=5)],
                     style={"columnCount":1})

app.run_server()
```

# Callbacks

```python
from dash.dependencies import Input, Output
```

```python
# Exemplo inutil de callback
ss = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
app = JupyterDash(__name__,external_stylesheets=ss)

app.layout = html.Div([
    html.H6("Header T6"),
    html.Div(["Insert: ",
            dcc.Input(id="input1", value="Type here...", type="text")]),
    html.Br(),
    html.Div(id="output1")
])

@app.callback(
    Output(component_id="output1", component_property="children"),
    Input(component_id="input1", component_property="value")
)

def update_output_div(input_value):
    return f"Saída: {input_value}"

app.run_server()
```

```python
# Slider
ss = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
app = JupyterDash(__name__,external_stylesheets=ss)

app.layout = html.Div([html.Label("Slider"),
                      dcc.Slider(0, 10, 1, value=5, id="slider"),
                      html.Br(),
                       html.Div(id="output1")],
                     style={"columnCount":1})

@app.callback(
    Output(component_id="output1", component_property="children"),
    Input(component_id="slider", component_property="value")
)

def update_output_div(input_value):
    return f"Saída: {input_value}"

app.run_server()
```

```python
from dash import dash_table
```

```python
ss = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
app = JupyterDash(__name__,external_stylesheets=ss)

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv')

app.layout = html.Div([
    dcc.Graph(id='graph-with-slider'),
    dcc.Slider(
        id='year-slider',
        min=df['year'].min(),
        max=df['year'].max(),
        value=df['year'].min(),
        marks={str(year): str(year) for year in df['year'].unique()},
        step=None
    ),
    html.H4(children='GdpPercap x LifeExp'),
    html.Div(id='table-wrapper',
             style={'width': 'auto','overflow-y':'scroll','max-width': '90vw', 'margin':'0 auto 50px'},     
             children=dash_table.DataTable(
                 id='table',
                 columns=[{"name":i, "id": i} for i in df.columns],
                 data=df.to_dict('records')
             )
    )               
])

@app.callback(
  Output('table-wrapper','children'),
  Input('year-slider','value'))
def update_figure(selected_year):
  filtered_df = df[df.year == selected_year]
  tab = dash_table.DataTable(
            id='table',
            columns=[{"name":i, "id": i} for i in filtered_df.columns],
            data=filtered_df.to_dict('records')
  )

  return tab



@app.callback(
  Output('graph-with-slider','figure'),
  Input('year-slider','value'))
def update_figure(selected_year):
  filtered_df = df[df.year == selected_year]

  fig = px.scatter(filtered_df, x='gdpPercap', y='lifeExp',
                   size='pop', color='continent', hover_name='country',
                   log_x=True, size_max=55)

  fig.update_layout(transition_duration=500)

  return fig

app.run_server(debug=True)
```
